# Model Building

```python
df.drop(columns=['animal name'],inplace=True)
```

```python
x=df.iloc[:,:-1]
y=df['type']
```

### Feature Scaling

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.75,random_state=1)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain,ytrain)
ypred = knn.predict(xtest)
ypred
```

```python
knn.score(xtrain,ytrain),knn.score(xtest,ytest)
```

```python
 print(classification_report(ytest,ypred))
```

# Hyperparameter Tunning for KNN

```python
trainac = []
testac = []

for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    ypred = knn.predict(xtest)
    trainac.append(knn.score(xtrain,ytrain))
    testac.append(knn.score(xtest,ytest))
```

```python
trainac
```

```python
testac
```

```python
plt.plot(range(1,30), trainac,linestyle = '-.', color = 'blue',label = 'Training Accuracy')
plt.plot(range(1,30),testac, linestyle = '--', color = 'green', label = 'Testing Accuracy')
plt.xlabel('K-value')
plt.ylabel('Accuracy')
plt.legend()
```

```python
knn = KNeighborsClassifier(n_neighbors=28)
knn.fit(xtrain,ytrain)
knn.score(xtrain,ytrain),knn.score(xtest,ytest)
```
