```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('my_movies.csv')
df.head()
```

```python
df.info()
```

```python
df = df.iloc[:,5:]
df
```

# Association Rule with 10% support and 70% confidence

```python
frequent_items = apriori(df, min_support=0.1, use_colnames=True)
frequent_items
```

```python
model = association_rules(frequent_items, metric='lift', min_threshold=0.7)
model
```

```python
model[model['lift']>1]
```

```python
plt.scatter(model['support'],model['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
