# Problem Statement

Use Random Forest to prepare a model on fraud data treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('Fraud_check .csv')
df.head()
```

```python
df.rename({'Marital.Status':'MS', 'Taxable.Income': 'Tax',
          'City.Population':'city_pop','Work.Experience':'Work'}, inplace=True, axis=1)
```

```python
df['Tax'] = df.Tax.map(lambda x: 'Risky' if x<=30000
                      else 'Good')

df.head()
```

```python
df['Tax'].nunique()
```

```python
df.info()
```

```python
le = LabelEncoder()

df['Undergrad'] = le.fit_transform(df['Undergrad'])
df['MS'] = le.fit_transform(df['MS'])
df['Tax'] = le.fit_transform(df['Tax'])
df['Urban'] = le.fit_transform(df['Urban'])
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.iloc[:,[0,1,3,4,5]]
y = df['Tax']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.75, random_state=143)
```

```python
rf = RandomForestClassifier(criterion='entropy')
rf.fit(xtrain,ytrain)
ypred = rf.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```

```python
print('Training accuracy: ', rf.score(xtrain,ytrain))
print('Testing Accuracy: ', rf.score(xtest,ytest))
```
